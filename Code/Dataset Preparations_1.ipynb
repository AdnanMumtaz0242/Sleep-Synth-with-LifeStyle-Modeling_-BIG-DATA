# Identify common and unique columns, and Standardize few Columns

```python
# Step 1: Identify common and unique columns
dataset_1_columns = set(dataset_1.columns)
dataset_2_columns = set(dataset_2.columns)

common_columns = dataset_1_columns.intersection(dataset_2_columns)
unique_to_dataset_1 = dataset_1_columns - dataset_2_columns
unique_to_dataset_2 = dataset_2_columns - dataset_1_columns

print("Common Columns:", common_columns)
print("Unique to Dataset 1:", unique_to_dataset_1)
print("Unique to Dataset 2:", unique_to_dataset_2)

# Step 2: Align the format of common columns
# Define a function to standardize columns
def standardize_column(df, column_name, dataset_number):
    if column_name == "Gender":
        df[column_name] = df[column_name].str.title()  # Ensure consistent capitalization
    elif column_name == "Age Group" and "Age" in df.columns:
        # Map numerical ages to age groups
        bins = [0, 18, 24, 34, 44, 54, 64, 100]
        labels = ["<18", "18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
        df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels)
    elif column_name == "Sleep Duration" and dataset_number == 2:
        # Convert numerical hours to "6+ Hours", "<6 Hours", etc.
        df["Sleep Duration"] = df["Sleep Duration"].apply(
            lambda x: "6+ Hours" if x >= 6 else "<6 Hours"
        )
    elif column_name == "BMI Category" and "BMI" in df.columns:
        # Map BMI to categories
        df["BMI Category"] = pd.cut(
            df["BMI"],
            bins=[0, 18.5, 24.9, 29.9, 100],
            labels=["Underweight", "Normal", "Overweight", "Obese"],
        )
    # Add other specific transformations here if needed
    return df

# Apply standardization to Dataset 2 for common columns
for column in common_columns:
    if column in dataset_2.columns:
        dataset_2 = standardize_column(dataset_2, column, dataset_number=2)

# Step 3: Add unique columns to Dataset 2 with NaN values or imputation
for column in unique_to_dataset_1:
    dataset_2[column] = None  # Add missing columns with None/NaN

# Step 4: Combine the datasets
combined_dataset = pd.concat([dataset_1, dataset_2], ignore_index=True)

# Step 5: Preview the combined dataset
print("Combined Dataset Shape:", combined_dataset.shape)
print("First few rows of Combined Dataset:")
```

# Modify sleep duration column to make the format same and make that 1 column in combine dataset

```python
# Define a function to standardize sleep duration into consistent categories
def standardize_sleep_duration(value):
    if isinstance(value, str) and "+" in value:  # Handle the first format
        return value.strip()  # Keep as is, e.g., "6+ Hours" or "4-6 Hours"
    elif isinstance(value, (int, float)):  # Handle numerical durations
        if value < 4:
            return "<4 Hours"
        elif 4 <= value < 6:
            return "4-6 Hours"
        elif value >= 6:
            return "6+ Hours"
    else:
        return "Unknown"  # Fallback for missing or unexpected values

# Standardize the "Sleep Duration" columns in the combined dataset
def combine_sleep_duration(data):
    # Ensure both columns are present
    col1 = "Sleep Duration (hrs/24hr)"
    col2 = "Sleep Duration"
    
    # Create a unified column
    data["Unified Sleep Duration"] = data[col1].combine_first(data[col2])
    data["Unified Sleep Duration"] = data["Unified Sleep Duration"].apply(standardize_sleep_duration)
    
    # Drop the old columns if no longer needed
    data.drop(columns=[col1, col2], inplace=True)
    
    return data

# Apply the transformation to the combined dataset
combined_dataset = combine_sleep_duration(combined_dataset)

# Preview the updated dataset
print("Updated Combined Dataset Shape:", combined_dataset.shape)
print("First few rows of Combined Dataset:")
```

# Modify Physical activity and exercise per days/week column and make that one 

```python
# Standardize column names (strip spaces, fix case sensitivity)
combined_dataset.columns = combined_dataset.columns.str.strip()

# Check if 'Physical Activity Level' exists in the dataset
if "Physical Activity Level" in combined_dataset.columns:
    # Define a mapping function to convert "Physical Activity Level" to "Exercise Days/Week"
    def convert_physical_activity_level(level):
        if level <= 30:
            return "0 Days"
        elif 31 <= level <= 60:
            return "1-2 Days"
        elif 61 <= level <= 90:
            return "3-4 Days"
        else:
            return "5+ Days"

    # Apply the conversion to the "Physical Activity Level" column
    combined_dataset["Exercise Days/Week"].fillna(
        combined_dataset["Physical Activity Level"].apply(convert_physical_activity_level),
        inplace=True
    )

    # Drop the "Physical Activity Level" column as it's no longer needed
    combined_dataset.drop(columns=["Physical Activity Level"], inplace=True)
else:
    print("Column 'Physical Activity Level' not found in the dataset.")

# Preview the updated dataset
print("Updated Combined Dataset Shape:", combined_dataset.shape)
print("First few rows of Combined Dataset:")
```

# Modify the BMI an d BMI category column to make that 1

```python
# Define a function to map BMI Category to approximate BMI values
def map_bmi_category(category):
    if isinstance(category, str):  # Check if the value is a string
        category = category.strip().lower()  # Normalize text for comparison
        if category in ["underweight"]:
            return 18.0  # Approximate value for Underweight
        elif category in ["normal", "normal weight"]:
            return 22.0  # Approximate value for Normal Weight
        elif category in ["overweight"]:
            return 27.0  # Approximate value for Overweight
        elif category in ["obese"]:
            return 32.0  # Approximate value for Obese
    return None  # Return None for invalid or missing categories

# For the first 109 subjects, use the "BMI" column
combined_dataset["Unified BMI"] = combined_dataset["BMI"]

# For the remaining subjects, fill in the mapped values from "BMI Category"
combined_dataset.loc[combined_dataset["Unified BMI"].isna(), "Unified BMI"] = (
    combined_dataset["BMI Category"].apply(map_bmi_category)
)

# Drop the original columns as they are now unified
combined_dataset.drop(columns=["BMI", "BMI Category"], inplace=True)

# Preview the updated dataset
print("Updated Combined Dataset Shape:", combined_dataset.shape)
print("First few rows of Combined Dataset:")
```

# Generating data for the missing Subjects of these columns "Stress Level	Blood Pressure	Heart Rate	Daily Steps" Using Gaussian sampling technique

```python
import numpy as np
import pandas as pd

# Load the combined dataset
combined_dataset = pd.read_csv("combined_dataset.csv")

# Separate the available data (subjects 110 onwards)
available_data = combined_dataset.iloc[109:]  # Assuming first 109 are missing

# Columns to generate data for
columns_to_generate = ["Stress Level", "Blood Pressure", "Heart Rate", "Daily Steps"]

# Handle missing or non-numeric values in numeric columns
numeric_columns = ["Stress Level", "Heart Rate", "Daily Steps"]
for column in numeric_columns:
    available_data[column] = pd.to_numeric(available_data[column], errors="coerce")  # Convert to numeric
    available_data = available_data.dropna(subset=[column])  # Drop rows with NaN in this column

# Generate data for numeric columns using Gaussian Sampling
for column in numeric_columns:
    mean = available_data[column].mean()
    std = available_data[column].std()

    # Ensure that min and max are not NaN
    if np.isnan(mean) or np.isnan(std):
        raise ValueError(f"Cannot compute mean or std for column '{column}'. Check data for issues.")

    min_val = available_data[column].min()
    max_val = available_data[column].max()

    generated_values = np.random.normal(mean, std, 109)  # Generate 109 values
    generated_values = np.clip(generated_values, min_val, max_val)  # Clip values to observed range
    combined_dataset.loc[:108, column] = generated_values

# Generate Blood Pressure (categorical-like) by random sampling
available_blood_pressure = available_data["Blood Pressure"].dropna().unique()
combined_dataset.loc[:108, "Blood Pressure"] = np.random.choice(available_blood_pressure, 109)

# Save the updated dataset
combined_dataset.to_csv("combined_dataset_with_generated_data.csv", index=False)

# Print confirmation
print("Data for the first 109 subjects has been generated and saved successfully!")
```

```python
# Remove the points value from these columns "Stress Level,	Heart Rate,	Daily Steps"
# List of columns to convert to integers
columns_to_convert = ["Stress Level", "Heart Rate", "Daily Steps"]

# Convert columns to integer format
for column in columns_to_convert:
    combined_dataset[column] = combined_dataset[column].round(0).astype(int)

# Save the updated dataset
combined_dataset.to_csv("combined_dataset_with_integers.csv", index=False)

# Print confirmation
print("Decimal values removed and converted to integers for specified columns!")
```

# Generate the data for the rest of the Missing columns 

```python
import pandas as pd
import numpy as np
import random

# Load your combined dataset
combined_dataset = pd.read_csv("combined_dataset.csv")

# Separate available data (first 109 subjects) and missing data (remaining subjects)
available_data = combined_dataset.iloc[:109]
missing_data = combined_dataset.iloc[109:]

# Categorical columns (like Language, Bedtime, etc.)
categorical_columns = [
    "Language", "Bedtime", "Wake-up Time", "Nap Duration", "Sleep Disturbances", "Sleep Medication"
]

# Numerical columns (like Height, Weight, etc.)
numerical_columns = [
    "Device Usage (hrs/day)", "Screen Time Before Sleep", "Height (cm)", "Weight (kg)", 
    "Sleep Onset Time", "Calculated Night Sleep Duration"
]

# 1. Generate Language Randomly from available languages (English, Urdu, Pashto)
languages = ["English", "Urdu", "Pashto"]
missing_data["Language"] = random.choices(languages, k=len(missing_data))

# 2. Generate Categorical Data (for columns like Bedtime, Sleep Disturbances, etc.) randomly based on available data distribution
for column in categorical_columns:
    # Ensure the column is categorical
    if available_data[column].dtype == 'object':
        # Generate random values based on the available data
        available_values = available_data[column].dropna().unique()
        missing_data[column] = random.choices(available_values, k=len(missing_data))

# 3. Generate Numerical Data using Gaussian Sampling (for columns like Height, Weight, etc.)
for column in numerical_columns:
    if available_data[column].dtype in ['int64', 'float64']:  # Ensure it's numeric
        mean = available_data[column].mean()
        std = available_data[column].std()
        # Generate data based on Gaussian distribution
        generated_values = np.random.normal(mean, std, len(missing_data))  # Generate the same number of values as missing data
        generated_values = generated_values.clip(min=available_data[column].min(), max=available_data[column].max())  # Clip values to observed range
        missing_data[column] = generated_values

# 4. Ensure the final dataset combines both available and generated data
combined_dataset.iloc[109:] = missing_data

# Save the updated combined dataset
combined_dataset.to_csv("combined_dataset_with_generated_data.csv", index=False)

# Print confirmation
print("Missing data generated and dataset saved as 'combined_dataset_with_generated_data.csv'!")
```

# Generate data for the leftovers columns 

```python
import pandas as pd
import random

# Load your combined dataset
combined_dataset = pd.read_csv("combined_dataset_with_generated_data.csv")

# Separate available data (first 109 subjects) and missing data (remaining subjects)
available_data = combined_dataset.iloc[:109]
missing_data = combined_dataset.iloc[109:]

# Columns for which we need to generate missing data
categorical_columns = [
    "Device Usage (hrs/day)", "Screen Time Before Sleep"
]

# 1. Generate Device Usage (hrs/day) for missing subjects based on the available distribution
for column in categorical_columns:
    # Check if the column exists in available_data
    if column in available_data.columns:
        # Get the distribution of values from available data
        available_values = available_data[column].dropna().value_counts(normalize=True)
        
        # Sample values for missing data based on the distribution
        missing_data[column] = random.choices(
            available_values.index, 
            weights=available_values.values, 
            k=len(missing_data)
        )

# 2. Ensure the final dataset combines both available and generated data
combined_dataset.iloc[109:] = missing_data

# Save the updated combined dataset
combined_dataset.to_csv("combined_dataset_with_generated_data.csv", index=False)

# Print confirmation
print("Missing data for 'Device Usage (hrs/day)' and 'Screen Time Before Sleep' generated and dataset saved!")
```

```python
import pandas as pd
import numpy as np

# Load your combined dataset
combined_dataset = pd.read_csv("combined_dataset_with_generated_data.csv")

# Columns that need modification
columns_to_round = ["Height (cm)", "Weight (kg)"]
column_to_one_decimal = "Calculated Night Sleep Duration"
column_to_add_random_float = "BMI"

# Handle NaN and infinite values in the specified columns before rounding
for column in columns_to_round:
    # Replace NaN with the column's median
    combined_dataset[column].fillna(combined_dataset[column].median(), inplace=True)
    # Replace infinite values with the column's mean
    combined_dataset[column].replace([np.inf, -np.inf], combined_dataset[column].mean(), inplace=True)
    # Round the values and convert to integers
    combined_dataset[column] = combined_dataset[column].round(0).astype(int)

# Handle NaN values and infinite values for "Calculated Night Sleep Duration"
combined_dataset[column_to_one_decimal].fillna(combined_dataset[column_to_one_decimal].median(), inplace=True)
combined_dataset[column_to_one_decimal].replace([np.inf, -np.inf], combined_dataset[column_to_one_decimal].mean(), inplace=True)
# Round to one decimal place
combined_dataset[column_to_one_decimal] = combined_dataset[column_to_one_decimal].round(1)

# Add a random floating point between 0 and 1 to the BMI column
combined_dataset[column_to_add_random_float] = combined_dataset[column_to_add_random_float] + np.random.rand(len(combined_dataset))

# Save the updated dataset
combined_dataset.to_csv("combined_dataset_with_generated_data.csv", index=False)

# Print confirmation
print("Data modifications complete and dataset saved as 'combined_dataset_with_modified_values.csv'!")
```
